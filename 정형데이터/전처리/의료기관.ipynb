```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 불러오기
raw = pd.read_csv('../../final_프젝_데이터/의료기관(구별)(2010년+이후)_20240731152428.csv',index_col=False)
```

```python
df = raw.copy()
```

```python
# '종로구' 데이터만 추출
jongno_df = df[df['자치구'] == '종로구']
# 종합병원, 병원, 의원, 요양병원 데이터만 선택
jongno_df = jongno_df[['종합병원 병원수', '병원 병원수', '의원 병원수', '요양병원 병원수']]
# 데이터 타입을 숫자로 변환 (데이터 값이 숫자로 표현되어야 함)
jongno_df = jongno_df.apply(pd.to_numeric, errors='ignore')
# 필요한 데이터 추출
data_for_plot = jongno_df.iloc[0].dropna()
```

```python
# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=data_for_plot.index, y=data_for_plot.values)
plt.xlabel('병원 종류')
plt.ylabel('병원 개수')
plt.title('종로구 병원 개수')
plt.rc('font', family='NanumBarunGothic')  # 폰트 설정
plt.xticks(rotation=45)
plt.show()
```
