```python
def map_address_column(dfnew, column_name, address_mapping):

    # 컬럼을 문자열로 변환
    dfnew[column_name] = dfnew[column_name].astype(str)

    # 매핑 적용
    dfnew[column_name] = dfnew[column_name].map(address_mapping)

    return dfnew
```
