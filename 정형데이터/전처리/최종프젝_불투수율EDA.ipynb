```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] =False
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 불러오기
impervious_df = pd.read_csv('/content/서울특별시_자치구별 불투수면적 현황_20201231.csv',encoding='euc-kr')  # 불투수율 데이터
```

```python
# 데이터 확인
impervious_df.head()
```

```python
# 기초 통계량 확인
impervious_df.describe()
```

```python
# 결측값 확인
print("\n불투수율 데이터 결측값 확인:")
impervious_df.isnull().sum()
```

```python
# 데이터 분포 시각화
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.rc('font', family='NanumBarunGothic')  # 폰트 설정
sns.histplot(impervious_df['불투수면적 비율(퍼센트)'], kde=True)
plt.title('불투수율 분포')
```

```python
# 자치구별 불투수면적 비율 시각화
plt.figure(figsize=(16, 10))
plt.rc('font', family='NanumBarunGothic')  # 폰트 설정

sns.barplot(x='불투수면적 비율(퍼센트)', y='자치구', data=impervious_df, palette='viridis')
plt.title('서울특별시 자치구별 불투수면적 비율')
plt.xlabel('불투수면적 비율(%)')
plt.ylabel('자치구')
plt.show()
```
