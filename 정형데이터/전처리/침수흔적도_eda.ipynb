```python
flood_counts = df.groupby(['자치구', '침수연도']).size().unstack(fill_value=0)


districts = flood_counts.index  # 자치구(행 이름)를 추출
years = flood_counts.columns[2:]  # 침수연도(열 이름)를 추출

# 두개로 나누기
mid_point = len(districts) // 2
districts_part1 = districts[:mid_point]
districts_part2 = districts[mid_point:]

# 각 부분의 x 위치를 설정
x_part1 = np.arange(len(districts_part1)) * 2
x_part2 = np.arange(len(districts_part2)) * 2
width = 0.35  # 막대 너비 설정 (기존보다 넓게)

# 두 개의 행을 가진 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 15), layout='constrained')

# 첫 번째 축에 그래프 그리기
multiplier = 0
for year in years:
    offset = width * multiplier
    rects = ax1.bar(x_part1 + offset, flood_counts.loc[districts_part1, year], width, label=f'{year}')
    ax1.bar_label(rects, padding=3)
    multiplier += 1

# 두 번째 축에 그래프 그리기
multiplier = 0
for year in years:
    offset = width * multiplier
    rects = ax2.bar(x_part2 + offset, flood_counts.loc[districts_part2, year], width, label=f'{year}')
    ax2.bar_label(rects, padding=3)
    multiplier += 1

# 첫 번째 축 커스터마이즈
ax1.set_ylabel('침수건수')
ax1.set_xticks(x_part1 + width, districts_part1, rotation=0)
ax1.legend(loc='upper left', ncols=3)
ax1.set_ylim(0, flood_counts.max().max() + 100)

# 두 번째 축 커스터마이즈
ax2.set_ylabel('침수건수')
ax2.set_xticks(x_part2 + width, districts_part2, rotation=0)
ax2.legend(loc='upper left', ncols=3)
ax2.set_ylim(0, flood_counts.max().max() + 100)

plt.show()
```

```python
import pandas as pd

# 데이터 로드
flood_data = pd.read_csv('침수흔적도_modified.csv')
rainfall_data = pd.read_csv('../강수량/서울시_일강수량.csv')

# 18년도 이후
flood_data = flood_data[flood_data['침수연도']>=2018]

# 침수 데이터에서 침수 시작 날짜를 문자열로 변환
flood_data['침수시작날짜'] = pd.to_datetime(flood_data['침수시작날짜'], format='%Y%m%d')

# 전날 날짜 열 추가
flood_data['전날'] = flood_data['침수시작날짜'] - pd.Timedelta(days=1)

# 침수 데이터에서 필요한 열만 추출
flood_relevant = flood_data[['자치구', '침수시작날짜', '전날', '침수수심']]

# 강수량 데이터의 날짜를 datetime 형식으로 변환
rainfall_data['날짜'] = pd.to_datetime(rainfall_data['날짜'], format='%Y%m%d')

# 당일 강수량 데이터 병합
merged_data = pd.merge(flood_relevant, rainfall_data[['날짜', '관측소명', '강수량']], left_on=['침수시작날짜', '자치구'], right_on=['날짜', '관측소명'], how='left')

# 전날 강수량 데이터 병합
merged_data = pd.merge(merged_data, rainfall_data[['날짜', '관측소명', '강수량']], left_on=['전날', '자치구'], right_on=['날짜', '관측소명'], how='left', suffixes=('_당일', '_전날'))

# 당일과 전날 강수량 중 더 큰 값 선택
merged_data['최고강수량'] = merged_data[['강수량_당일', '강수량_전날']].max(axis=1)

# NaN 값을 포함하는 행 제거
merged_data_clean = merged_data.dropna(subset=['최고강수량', '침수수심'])

# 침수수심을 날짜와 자치구별로 최대값으로 집계
max_flood_depth = merged_data_clean.groupby(['자치구', '침수시작날짜'])['침수수심'].max().reset_index()

# 최고 강수량과 병합
final_data = pd.merge(max_flood_depth, merged_data_clean[['자치구', '침수시작날짜', '최고강수량']].drop_duplicates(), on=['자치구', '침수시작날짜'])
```
