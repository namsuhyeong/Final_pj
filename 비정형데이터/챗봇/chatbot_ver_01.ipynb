```python
# Create a Python file for the Streamlit app
with open('app.py', 'w') as f:
    f.write("""
import streamlit as st
import pandas as pd
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration
from difflib import SequenceMatcher
import torch
import re

# KoBART 모델 로드
tokenizer = PreTrainedTokenizerFast.from_pretrained("gogamza/kobart-base-v1")
model = BartForConditionalGeneration.from_pretrained("gogamza/kobart-summarization")

# 데이터 로드
@st.cache
def load_data():
    return pd.read_csv('NEW_뉴스_크롤링_data.csv')

news_data = load_data()

# 기사를 요약하는 함수
def summarize_article(article_text):
    if len(article_text.strip()) == 0:
        return "요약할 내용이 없습니다."

    # 요약 수행 (min_length 제거)
    inputs = tokenizer([article_text], max_length=1024, return_tensors="pt", truncation=True)
    summary_ids = model.generate(inputs.input_ids, max_length=150, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # 중복 문구 제거 로직 추가
    summary = remove_repeated_phrases(summary)

    return summary if len(summary.strip()) > 0 else "요약할 내용이 충분하지 않습니다."

# 중복된 문구 제거 함수
def remove_repeated_phrases(text):
    words = text.split()
    unique_words = []
    for word in words:
        if len(unique_words) < 1 or word != unique_words[-1]:
            unique_words.append(word)
    return ' '.join(unique_words)

# 특정 지역이 제목이나 내용에 포함된 뉴스를 필터링하는 함수 (가중치 적용)
def get_news_by_keywords_with_weight(keywords, data, num_results=5, exclude_ids=None):
    pattern = '|'.join(keywords)
    filtered_news = data[
        data['제목'].str.contains(pattern, na=False, flags=re.IGNORECASE) |
        data['content'].str.contains(pattern, na=False, flags=re.IGNORECASE)
    ]

    if exclude_ids is not None:
        filtered_news = filtered_news[~filtered_news.index.isin(exclude_ids)]  # 이미 표시된 기사를 제외하도록 수정했습니다.

    filtered_news['weight'] = filtered_news['제목'].apply(
        lambda x: sum([1 for keyword in keywords if keyword in x])
    )
    filtered_news = filtered_news.sort_values(
        by=['weight', '날짜'], ascending=[False, False]
    )

    return filtered_news.head(num_results)

# 질문에서 지역 정보를 추출하는 함수
def extract_regions_from_question(question):
    cities = ['서울시']
    districts = [
        '종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구',
        '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구', '강서구',
        '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구'
    ]

    regions = []
    for city in cities:
        if city in question:
            regions.append(city.replace('시', ''))

    for district in districts:
        if district in question:
            regions.append(district)

    return regions

# 질문에서 관련 키워드를 추출하는 함수
def extract_keywords_from_question(question):
    keywords = ['침수', '홍수', '재해', '폭우', '강우량']
    extracted_keywords = []

    for keyword in keywords:
        if keyword in question:
            extracted_keywords.append(keyword)

    return extracted_keywords

# Streamlit 앱 구현
def main():
    st.title("침수 뉴스 챗봇")
    st.write("원하시는 질문을 입력하시면, 관련 뉴스를 제공해드립니다.")

    session = st.session_state

    if 'last_query' not in session:
        session.last_query = None
        session.last_results = None
        session.last_displayed_ids = []  # 마지막으로 표시된 기사들의 ID를 저장하기 위한 리스트를 추가했습니다.
        session.awaiting_region = False

    user_input = st.text_input('질문을 입력하세요:')

    if st.button('검색'):
        if user_input:
            regions = extract_regions_from_question(user_input)
            keywords = extract_keywords_from_question(user_input)
            search_terms = regions + keywords
            results = get_news_by_keywords_with_weight(search_terms, news_data, num_results=5)
            session.last_query = search_terms
            session.last_results = results
            session.last_displayed_ids = results.index.tolist()  # 표시된 기사들의 ID를 저장합니다.

            if results.empty:
                st.write("{0}와 관련된 뉴스를 찾을 수 없습니다.".format(', '.join(search_terms)))
            else:
                st.write("{0}와 관련된 대표적인 뉴스 기사 5개입니다:".format(', '.join(search_terms)))
                for index, row in results.iterrows():
                    summary = summarize_article(row['content'])  # 기사 요약
                    st.markdown(f"### **{row['제목']} ({row['날짜']})**")  # 제목을 큰 글씨로, 굵게 표시
                    st.markdown(f"{summary}")  # 요약된 내용을 표시
                    st.markdown(f"[기사 원문 보러가기]({row['링크']})")  # 링크 추가

                if len(results) == 5:
                    st.session_state.show_more = True  # "더 보여줘" 버튼 표시를 위한 상태 변수를 추가했습니다.

    if session.last_results is not None and st.session_state.get('show_more', False) and st.button('더 보여줘', key='show_more_btn'):
        additional_results = get_news_by_keywords_with_weight(session.last_query, news_data, num_results=20, exclude_ids=session.last_displayed_ids)  # 이미 표시된 기사들을 제외하도록 `exclude_ids` 파라미터를 전달합니다.

        if additional_results.empty:
            st.write("더 이상 관련 기사가 존재하지 않습니다.")
            st.session_state.show_more = False  # 더 이상 버튼을 표시하지 않도록 상태를 업데이트합니다.
        else:
            session.last_displayed_ids.extend(additional_results.index.tolist())  # 새로 표시된 기사들의 ID를 추가합니다.
            st.write("추가로 관련된 뉴스 기사 20개를 보여드립니다:")
            for index, row in additional_results.iterrows():
                summary = summarize_article(row['content'])  # 기사 요약
                st.markdown(f"### **{row['제목']} ({row['날짜']})**")  # 제목을 큰 글씨로, 굵게 표시
                st.markdown(f"{summary}")  # 요약된 내용을 표시
                st.markdown(f"[기사 원문 보러가기]({row['링크']})")  # 링크 추가

            if len(additional_results) == 20:
                st.button('더 보여줘', key=f'show_more_btn_{len(session.last_displayed_ids)}')  # 각 버튼에 고유한 키를 추가했습니다.
            else:
                st.write("더 이상 관련 기사가 존재하지 않습니다.")
                st.session_state.show_more = False  # 더 이상 버튼을 표시하지 않도록 상태를 업데이트합니다.

if __name__ == "__main__":
    main()

""")
```

```python
# 데이터 로드
file_path = 'NEW_뉴스_크롤링_data.csv'
data = pd.read_csv(file_path)

# 데이터 확인
data.head()
```

```python
from sentence_transformers import SentenceTransformer

# 임베딩 모델 로드 (Sentence-BERT)
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# 텍스트 데이터를 임베딩 벡터로 변환
def embed_text(text):
    return model.encode(text).astype('float32')

# 데이터의 제목과 내용을 결합한 후 임베딩 벡터로 변환
data['embedding'] = data.apply(lambda row: embed_text(f"{row['제목']} {row['content']}"), axis=1)
```

```python
import numpy as np

# 임베딩 벡터를 NumPy 배열로 변환
embeddings = np.stack(data['embedding'].values)

# 데이터 인덱스를 함께 저장하고 싶다면 아래처럼 활용할 수 있습니다.
indices = data.index.values  # 데이터의 인덱스를 함께 저장하기 위해
```

```python
# 임베딩 벡터를 저장 (.npy 파일로)
np.save('embeddings.npy', embeddings)

# 인덱스도 함께 저장 (선택 사항)
np.save('indices.npy', indices)
```

```python
import pickle

# 데이터프레임을 Pickle 파일로 저장
with open('data_with_embeddings.pkl', 'wb') as f:
    pickle.dump(data, f)

# Pickle 파일에서 불러오기
with open('data_with_embeddings.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
# 3. Streamlit 앱 실행
# BM25 코드 (이걸 돌려야함)
with open('app.py', 'w') as f:
    f.write("""
import pandas as pd
import pickle
import streamlit as st
from rank_bm25 import BM25Okapi
from nltk.tokenize import word_tokenize
import nltk

# NLTK 데이터 다운로드 (토크나이저를 사용하기 위해 필요)
nltk.download('punkt')

# 데이터 로드
with open('news_data.pkl', 'rb') as f:
    data = pickle.load(f)

# 제목과 내용을 결합한 텍스트 생성
data['text'] = data['제목'] + " " + data['content']

# 텍스트를 토큰화 (모든 문서에 대해)
tokenized_corpus = [word_tokenize(text) for text in data['text']]

# BM25 객체 생성 (모든 문서에 대해)
bm25 = BM25Okapi(tokenized_corpus)

# 침수 사건 뉴스 필터링
def filter_flood_incidents(news_df):
    keywords = ['침수', '피해', '홍수', '물난리', '재해','기상','예보']
    flood_news = news_df[news_df['제목'].str.contains('|'.join(keywords), case=False, na=False)]
    other_news = news_df[~news_df['제목'].str.contains('|'.join(keywords), case=False, na=False)]
    return flood_news, other_news

# 유사한 뉴스를 검색하는 함수
def search_similar_news(query, top_k=10):
    # 쿼리 토큰화
    tokenized_query = word_tokenize(query)

    # BM25 점수 계산 및 상위 문서 반환
    top_n = bm25.get_top_n(tokenized_query, data['text'], n=top_k)

    # 상위 문서를 데이터프레임으로 반환
    results = data[data['text'].isin(top_n)]

    # 침수 사건 뉴스와 기타 뉴스로 분리
    flood_news, other_news = filter_flood_incidents(results)
    return flood_news, other_news

# Streamlit 앱 구현
def main():
    st.title("침수 예측 뉴스 챗봇")
    st.write("서울시 구별 침수 관련 뉴스를 찾아드립니다.")

    # 서울시의 구 목록
    seoul_districts = [
        "종로구", "중구", "용산구", "성동구", "광진구", "동대문구", "중랑구",
        "성북구", "강북구", "도봉구", "노원구", "은평구", "서대문구", "마포구",
        "양천구", "강서구", "구로구", "금천구", "영등포구", "동작구", "관악구",
        "서초구", "강남구", "송파구", "강동구"
    ]

    # 사용자가 선택할 수 있는 드롭다운 메뉴
    selected_district = st.selectbox("침수 관련 뉴스를 볼 지역을 선택하세요:", seoul_districts)

    # 세션 상태 초기화
    if 'flood_news' not in st.session_state:
        st.session_state.flood_news = pd.DataFrame()
    if 'other_news' not in st.session_state:
        st.session_state.other_news = pd.DataFrame()

    if st.button("검색"):
        # 사용자가 선택한 구와 관련된 뉴스를 검색
        query = selected_district + " 침수"
        flood_news, other_news = search_similar_news(query)

        # 세션 상태에 뉴스 저장
        st.session_state.flood_news = flood_news
        st.session_state.other_news = other_news

    # 검색 결과가 존재할 때만 뉴스 표시
    if not st.session_state.flood_news.empty:
        st.write(f"{selected_district}와 관련된 침수 사건 뉴스 기사:")
        for idx, row in st.session_state.flood_news.iterrows():
            st.markdown(f"### **{row['제목']}**")
            st.write(f"{row['content']}")
            st.write(f"날짜: {row['날짜']}")
            st.markdown(f"[기사 원문 보러가기]({row['링크']})")

    # "더 많은 뉴스 보기" 버튼
    if not st.session_state.other_news.empty and st.button("더 많은 뉴스 보기"):
        st.write(f"{selected_district}와 관련된 기타 뉴스 기사:")
        for idx, row in st.session_state.other_news.iterrows():
            st.markdown(f"### **{row['제목']}**")
            st.write(f"{row['content']}")
            st.write(f"날짜: {row['날짜']}")
            st.markdown(f"[기사 원문 보러가기]({row['링크']})")

if __name__ == "__main__":
    main()
    """)
```
