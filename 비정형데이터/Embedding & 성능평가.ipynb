```python
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient
from qdrant_client.models import PointStruct, VectorParams, Distance
import openai

encoder = SentenceTransformer('all-MiniLM-L6-v2')

# OpenAI API 키 설정
openai.api_key = ''  # 자신의 OpenAI API 키로 교체하세요

def get_openai_embedding(text):
    response = openai.Embedding.create(
        model="text-embedding-ada-002",  # OpenAI의 모델 예시
        input=text
    )
    return response['data'][0]['embedding']

def create_qdrant_collection(collection_name, model_type, dictionary):
    if model_type == 'openai':
        encoder = get_openai_embedding
        vector_size = 1536  # OpenAI's text-embedding-ada-002 vector size
    elif model_type == 'sentence_transformer':
        encoder = SentenceTransformer('all-MiniLM-L6-v2')
        vector_size = encoder.get_sentence_embedding_dimension()
    else:
        raise ValueError("Invalid model_type. Choose 'openai' or 'sentence_transformer'.")

    qdrant = QdrantClient(
        url='https://119f2970-ba35-42b1-9e56-5826b62fd428.europe-west3-0.gcp.cloud.qdrant.io',
        api_key=""
    )

    qdrant.recreate_collection(
        collection_name=collection_name,
        vectors_config=VectorParams(
            size=vector_size,
            distance=Distance.COSINE
        )
    )

    points = []
    for idx, doc in enumerate(dictionary):
        if model_type == 'openai':
            encoded_vector = get_openai_embedding(doc['content'])
        elif model_type == 'sentence_transformer':
            encoded_vector = encoder.encode(doc['제목'] + " " + doc['content']).tolist()

        # 벡터의 길이가 0이 아니어야 업로드
        if len(encoded_vector) > 0:
            points.append(
                PointStruct(
                    id=idx,
                    vector=encoded_vector,
                    payload=doc
                )
            )
        else:
            print(f"Warning: Empty vector for document with id {idx}. Skipping this document.")

    # 업로드
    if points:
        qdrant.upload_points(
            collection_name=collection_name,
            points=points
        )
    else:
        print("No valid points to upload.")
```

```python
from qdrant_client import QdrantClient
from qdrant_client.models import PointStruct, VectorParams, Distance
import openai

# OpenAI API 키 설정
openai.api_key = ''  # 자신의 OpenAI API 키로 교체하세요

# OpenAI 임베딩 모델을 사용하여 텍스트 임베딩 벡터 생성하는 함수
def get_openai_embedding(text, model_name):
    response = openai.Embedding.create(
        model=model_name,  # OpenAI 임베딩 모델 이름
        input=text
    )
    return response['data'][0]['embedding']

# Qdrant 컬렉션 생성 함수
def create_qdrant_collection(collection_name, model_name, dictionary):
    # 모델 타입에 따라 벡터 크기 설정
    if model_name == 'text-embedding-ada-002':
        vector_size = 1536  # OpenAI's text-embedding-ada-002 벡터 크기
    elif model_name == 'text-embedding-3-small':
        vector_size = 1536  # 예시: 모델에 따라 실제 벡터 크기로 설정
    elif model_name == 'text-embedding-3-large':
        vector_size = 3072  # 예시: 모델에 따라 실제 벡터 크기로 설정
    else:
        raise ValueError("Invalid model name. Choose from 'text-embedding-ada-002', 'text-embedding-3-small', or 'text-embedding-3-large'.")

    # Qdrant 클라이언트 초기화
    qdrant = QdrantClient(
        url='https://119f2970-ba35-42b1-9e56-5826b62fd428.europe-west3-0.gcp.cloud.qdrant.io',
        api_key=""
    )

    # Qdrant 컬렉션 생성 또는 재생성
    qdrant.recreate_collection(
        collection_name=collection_name,
        vectors_config=VectorParams(
            size=vector_size,
            distance=Distance.COSINE
        )
    )

    points = []
    for idx, doc in enumerate(dictionary):
        text = doc['제목'] + " " + doc['content']
        # 선택한 모델에 따라 임베딩 생성
        encoded_vector = get_openai_embedding(text, model_name)

        # 벡터의 길이가 0이 아니어야 업로드
        if len(encoded_vector) > 0:
            points.append(
                PointStruct(
                    id=idx,
                    vector=encoded_vector,
                    payload=doc
                )
            )
        else:
            print(f"Warning: Empty vector for document with id {idx}. Skipping this document.")

    # Qdrant에 벡터 업로드
    if points:
        qdrant.upload_points(
            collection_name=collection_name,
            points=points
        )
    else:
        print("No valid points to upload.")

# # 예시 사용법
# documents = [
#     {"제목": "Example Title 1", "content": "Example content 1"},
#     {"제목": "Example Title 2", "content": "Example content 2"}
# ]

# # 각 모델을 사용해 컬렉션 생성
# create_qdrant_collection('news_bot_openai_ada2_500_26', 'text-embedding-ada-002', dictionary)
# create_qdrant_collection('news_bot_openai_small3_500_26', 'text-embedding-3-small', dictionary)
# create_qdrant_collection('news_bot_openai_large3_500_26', 'text-embedding-3-large', dictionary)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from rank_bm25 import BM25Okapi
import spacy
import openai

# OpenAI API 키 설정
openai.api_key = ''  # OpenAI API 키를 직접 입력

# spaCy 모델 로드
nlp = spacy.load('en_core_web_sm')

def spacy_tokenize(text):
    doc = nlp(text)
    return [token.text for token in doc]

# 문서 내용
doc_contents = [doc['content'] for doc in documents]

# 쿼리
query = "[물순환] 제주 수자원 보호 '자연적 빗물 흐름 관리' 과제 대두"

# 1. BM25 검색
def bm25_search(query, documents):
    tokenized_docs = [spacy_tokenize(doc) for doc in documents]
    tokenized_query = spacy_tokenize(query)

    bm25 = BM25Okapi(tokenized_docs)
    scores = bm25.get_scores(tokenized_query)
    return scores

bm25_scores = bm25_search(query, doc_contents)
print("BM25 Scores:", bm25_scores)

# 2. 벡터 검색 (TF-IDF)
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(doc_contents + [query])
tfidf_scores = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])
print("TF-IDF Cosine Similarity Scores:", tfidf_scores.flatten())


# OpenAI 임베딩 함수
def get_openai_embedding(text):
    response = openai.Embedding.create(input=text, model="text-embedding-ada-002")
    return response['data'][0]['embedding']

# OpenAI 임베딩을 사용한 문서 벡터화
doc_embeddings = np.array([get_openai_embedding(doc) for doc in doc_contents])

# OpenAI 임베딩을 사용한 쿼리 벡터화
query_embedding = np.array(get_openai_embedding(query)).reshape(1, -1)

cosine_sim = cosine_similarity(query_embedding, doc_embeddings)
print("vector search Cosine Similarity Scores:", cosine_sim.flatten())

# 4. 앙상블 리트리버
def ensemble_retriever(bm25_scores, tfidf_scores, sentence_scores):
    # 정규화
    norm_bm25 = (bm25_scores - np.min(bm25_scores)) / (np.max(bm25_scores) - np.min(bm25_scores))
    norm_tfidf = (tfidf_scores.flatten() - np.min(tfidf_scores.flatten())) / (np.max(tfidf_scores.flatten()) - np.min(tfidf_scores.flatten()))
    norm_sentence = (sentence_scores.flatten() - np.min(sentence_scores.flatten())) / (np.max(sentence_scores.flatten()) - np.min(sentence_scores.flatten()))

    # 단순 평균 앙상블
    ensemble_scores = (norm_bm25 + norm_tfidf + norm_sentence) / 3
    return ensemble_scores

ensemble_scores = ensemble_retriever(bm25_scores, tfidf_scores, cosine_sim)
print("Ensemble Scores:", ensemble_scores)

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame({
    'Document ID': range(len(doc_contents)),
    'BM25 Scores': bm25_scores,
    'TF-IDF Cosine Similarity Scores': tfidf_scores.flatten(),
    'Qdrant db search score': cosine_sim.flatten(),
    'Ensemble Scores': ensemble_scores
})

print(results_df)
```

```python
import pandas as pd
from PIL import Image
import os
import base64
from qdrant_client import QdrantClient
from sentence_transformers import SentenceTransformer
import nest_asyncio
import openai
from rank_bm25 import BM25Okapi
import re
import time
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np




# 환경 변수 설정 (OpenAI API Key)
os.environ["OPENAI_API_KEY"] = ""

# OpenAI API 설정
openai.api_key = os.environ["OPENAI_API_KEY"]

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# 동-구 매핑 데이터 예시
df = pd.read_csv('/content/drive/MyDrive/최종프로젝트/data/dong_to_gu.csv')
dong_to_gu = dict(zip(df['법정동'], df['자치구']))
gu_list = df['자치구'].unique()

# Qdrant 클라이언트 및 SentenceTransformer 설정
def get_openai_embedding(text):
    response = openai.Embedding.create(
        model="text-embedding-3-small",  # OpenAI의 모델 예시
        input=text
    )
    return response['data'][0]['embedding']

qdrant = QdrantClient(
    url='https://119f2970-ba35-42b1-9e56-5826b62fd428.europe-west3-0.gcp.cloud.qdrant.io',  # Qdrant Cloud의 URL
    api_key=""  # Qdrant API 키
)

# BM25 필터링 추가
keywords = ['침수', '피해', '홍수', '물난리', '재해', '기상']  # 필터링할 키워드 목록

def apply_ensemble_retriever(docs, query_vector):
    """BM25 점수와 OpenAI 임베딩 유사성을 결합한 Ensemble Retriever 적용"""

    # BM25 필터링
    tokenized_docs = [doc['content'].split() for doc in docs]
    bm25 = BM25Okapi(tokenized_docs)
    tokenized_keywords = keywords
    bm25_scores = bm25.get_scores(tokenized_keywords)

    # 이미 Qdrant에서 가져온 임베딩을 사용하여 코사인 유사성 계산
    doc_embeddings = np.array([doc['embedding'] for doc in docs])
    cosine_scores = cosine_similarity([query_vector], doc_embeddings)[0]

    # 정규화
    norm_bm25 = (bm25_scores - np.min(bm25_scores)) / (np.max(bm25_scores) - np.min(bm25_scores))
    norm_cosine = (cosine_scores - np.min(cosine_scores)) / (np.max(cosine_scores) - np.min(cosine_scores))

    # 가중치 설정
    bm25_weight = 0.1
    cosine_weight = 0.9

    # 가중 합 계산
    combined_scores = bm25_weight * norm_bm25 + cosine_weight * norm_cosine

    # 점수에 따라 문서 정렬
    ranked_docs = [doc for _, doc in sorted(zip(combined_scores, docs), key=lambda x: x[0], reverse=True)]

    return ranked_docs

def search_qdrant(query_text):
    """Qdrant에서 관련 데이터를 검색하는 함수"""
    query_vector = model.encode(query_text)
    search_result = qdrant.search(
        collection_name="news_bot_huggingface_0904_final",
        query_vector=query_vector,
        limit=5,
        with_vectors=True
    )

    search_results = []
    for point in search_result:
        payload = point.payload
        search_results.append({
            "title": payload['제목'],
            "link": payload['링크'],
            "date": payload['날짜'],
            "source": payload['언론사'],
            "content": payload['content'],
            "embedding": point.vector  # 임베딩 벡터를 함께 반환
        })

    # Ensemble Retriever 적용
    ranked_results = apply_ensemble_retriever(search_results, query_vector)

    context_text = ""
    for result in ranked_results:
        context_text += f"제목: {result['title']}\n링크: {result['link']}\n날짜: {result['date']}\n출처: {result['source']}\n내용: {result['content']}\n\n"

    return context_text, ranked_results

# 'None'이 나오는 문제 해결 및 제목/내용을 한글로 변경
def get_openai_response_1(context_text, user_query, dong_name=None, gu_name=None, article_list=None):
    """dong_name과 gu_name의 None 처리 및 출력 형식 수정"""

    # 만약 동 이름이 없고 자치구 이름만 있을 경우
    if dong_name is None:
        location_name = gu_name  # 자치구 이름만 표시
    else:
        location_name = f"{dong_name}({gu_name})"  # 동과 자치구 이름 모두 표시

    # 침수 관련 정보를 사용자 쿼리에 맞춰 제공
    if "침수" in user_query:
        specific_info = (
            f"{location_name}이 침수 시 아래와 같은 조치를 취하세요:\n"
            f"  - 물이 집 안으로 들어올 수 있는 곳을 차단하세요.\n"
            f"  - 전자기기 및 귀중품을 안전한 곳으로 옮기세요.\n"
            f"  - 하천이나 도로가 범람할 위험이 있으니 피하세요.\n"
            f"  - 안전한 장소로 대피하고, 비상 연락처를 확인하세요.\n"
        )
    elif "지하건물" in user_query or "지하 피해" in user_query:
        specific_info = (
            f"{location_name}의 지하 건물 피해를 예방하기 위해 다음을 참고하세요:\n"
            f"  - 지하실의 배수펌프가 정상 작동하는지 확인하세요.\n"
            f"  - 중요한 물품은 지하에 두지 말고 높은 곳으로 이동시키세요.\n"
            f"  - 침수 예보 시 지하 공간 사용을 자제하세요.\n"
        )
    else:
        specific_info = (
            f"{location_name}에 대한 추가 정보가 필요할 경우, 관련 기사를 확인하거나 구청에 문의하세요.\n"
        )

    try:
        # OpenAI 모델을 사용하여 응답 생성
        messages = [
            {
                "role": "system",
                "content": (
                    "You are a helpful assistant. Answer in Korean. "
                    "Please focus on the specific user query and provide detailed, relevant information. "
                    "Ensure that any articles are summarized briefly with a link."

                    )
            },
            {
                "role": "user",
                "content": user_query
            }
        ]

        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",
            messages=messages,
            max_tokens=1500,
            n=1,
            stop=None,
            temperature=0.7
        )
        openai_result = response.choices[0].message['content'].strip()

        # 응답을 형식에 맞게 정리하여, 기사가 위로 올라오고 추가 정보가 아래로 배치되게 수정
        return format_response(location_name, gu_name, article_list, openai_result, specific_info)

    except Exception as e:
        return f"질문 처리 중 오류가 발생했습니다: {e}"

def get_openai_response_2(context_text, user_query, article_list):
    """dong_name, gu_name이 없는 경우의 질문 처리"""

    system_prompt = (
        "You are a helpful assistant that answers questions about flood risks and regional information in Korea. "
        "You provide clear, concise, and relevant answers based on the given information. "
        "If the response includes a specific location (within Seoul), make sure to mention that location and related data, such as flood risk, green areas, or other relevant information. "
        "Always respond in Korean. Format the response with Markdown, use bullet points for lists, and ensure that if any articles are referenced, they are formatted as links with a brief summary."
    )

    user_prompt = (
        f"{user_query}\n\n"
        f"뉴스 기사에서 발견된 관련 정보:\n"
        f"{context_text}\n\n"
        f"#### 관련 뉴스 기사 목록:\n{article_list}\n\n"
        "해당 기사와 관련된 침수 지역 정보가 있으면 제공해주세요."
    )

    try:
        messages = [
            {
                "role": "system",
                "content": system_prompt
            },
            {
                "role": "user",
                "content": user_prompt
            }
        ]

        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=messages,
            max_tokens=1500,
            n=1,
            stop=None,
            temperature=0.7
        )
        openai_result = response.choices[0].message['content'].strip()

        # 응답을 형식에 맞게 정리하여 반환
        return openai_result

    except Exception as e:
        return f"질문 처리 중 오류가 발생했습니다: {e}"


# 'Title', 'Content'를 '제목', '요약 내용'으로 변경
def format_articles(articles):
    """Qdrant에서 불러온 기사를 한글로 정리, 제목과 요약 내용 사이에 명확한 줄바꿈 추가"""
    formatted_articles = ""
    for article in articles:
        formatted_articles += (
            f"**제목:** [{article['title']}]({article['link']})  \n"  # 줄바꿈을 위한 두 개의 공백 추가
            f"**본문 내용:** {article['content'][:200]}...  \n\n"  # 줄바꿈을 위한 두 개의 공백 추가
        )
    return formatted_articles


def format_response(location_name, gu_name, article_list, context_text, specific_info):
    """AI의 응답을 형식화하는 함수"""

    return (
        f"{location_name}\n"
        f"- 해당 지역에서 침수와 관련된 여러 예측 결과가 있습니다.\n"
        f"- 왼쪽 지도는 {location_name}이 속해 있는 {gu_name}의 침수 예측 지도입니다.\n"
        f"- {location_name} 정보는 다음과 같습니다:\n\n"
        f"{specific_info}\n\n"  # 사용자 쿼리와 관련된 정보 제공
        f"{context_text}\n\n"  # AI의 추가적인 답변도 기사보다 먼저 출력
        f"#### {location_name} 관련 뉴스 기사:\n{article_list}\n\n"
    )


dong_name = '역삼동'
gu_name = '강남구'
user_query = f'{gu_name} 침수'




context_text, search_results = search_qdrant(user_query)
article_list = format_articles(search_results)


# OpenAI API를 사용하여 응답 생성
ai_responses = get_openai_response_1(context_text, user_query,dong_name, gu_name, article_list)

ai_responses
```

```python
import numpy as np
from collections import Counter
from nltk import ngrams

# 후보 단어 리스트
candidate = words_list

# 참조 문장
reference_list = [doc['content'].split() for doc in documents]

# n-gram 카운트
def simple_count(tokens, n):
    return Counter(ngrams(tokens, n))

# 정밀도 계산
def count_clip(candidate, reference_list, n):
    ca_cnt = simple_count(candidate, n)
    max_ref_cnt_dict = dict()

    for ref in reference_list:
        ref_cnt = simple_count(ref, n)
        for n_gram in ref_cnt:
            if n_gram in max_ref_cnt_dict:
                max_ref_cnt_dict[n_gram] = max(ref_cnt[n_gram], max_ref_cnt_dict[n_gram])
            else:
                max_ref_cnt_dict[n_gram] = ref_cnt[n_gram]

    return {
        n_gram: min(ca_cnt.get(n_gram, 0), max_ref_cnt_dict.get(n_gram, 0)) for n_gram in ca_cnt
    }


def modified_precision(candidate, reference_list, n):
    clip_cnt = count_clip(candidate, reference_list, n)
    total_clip_cnt = sum(clip_cnt.values())  # 분자

    cnt = simple_count(candidate, n)
    total_cnt = sum(cnt.values())
    # 분모
    if total_cnt == 0:  # 분모가 0이 되는 것을 방지
        total_cnt = 1

    return (total_clip_cnt / total_cnt)

# 가장 근접한 참조 문장의 길이를 찾는 함수
def closest_ref_length(candidate, reference_list):
    ca_len = len(candidate)
    ref_lens = (len(ref) for ref in reference_list)
    closest_ref_len = min(ref_lens, key=lambda ref_len: (abs(ref_len - ca_len), ref_len))
    return closest_ref_len


def brevity_penalty(candidate, reference_list):
    ca_len = len(candidate)
    ref_len = closest_ref_length(candidate, reference_list)

    # 짧은 길이의 문장에 패널티 적용
    if ca_len > ref_len:
        return 1
    elif ca_len == 0:
        return 0
    else:
        return np.exp(1 - ref_len / ca_len)

# BLEU 점수 계산
def bleu_score(candidate, reference_list, weights=[0.1, 0.1, 0.1, 0.1]):
    bp = brevity_penalty(candidate, reference_list)
    p_n = [modified_precision(candidate, reference_list, n=n) for n, _ in enumerate(weights, start=1)]
    score = np.sum([w_i * np.log(p_i) if p_i != 0 else 0 for w_i, p_i in zip(weights, p_n)])
    return bp * np.exp(score)

# 여러 가중치 설정 및 BLEU 점수 계산
weights_list = [
    [0.1, 0.1, 0.1, 0.1],     # 균등 가중치
    [0.5, 0.2, 0.2, 0.1],     # 1-그램 중요도 높음
    [0.1, 0.3, 0.3, 0.3],     # 2-그램, 3-그램 중요도 높음
    [0.2, 0.2, 0.2, 0.4],     # 4-그램 중요도 높음
    [0.4, 0.3, 0.2, 0.1]      # 1-그램 중요도 높음
]

for weights in weights_list:
    bleu = bleu_score(candidate, reference_list, weights=weights)
    print(f"BLEU Score with weights {weights}: {bleu}")
```
