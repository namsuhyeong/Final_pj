```python
import pandas as pd
import numpy as np

# CSV 파일 로드
file_path = 'NEW_뉴스_크롤링_data.csv'
news_data = pd.read_csv(file_path)

# "침수" 관련 키워드
flood_keywords = ['침수', '피해', '홍수', '물난리', '재해', '기상', '예보']

# 정치 관련 키워드
political_keywords = ["대통령", "의원", "시장", "정당", "청와대", "정부", "정치인", "총리"]

# 침수 관련 키워드가 포함된 기사만 남기기
flood_related = news_data[
    news_data['제목'].str.contains('|'.join(flood_keywords), case=False) |
    news_data['content'].str.contains('|'.join(flood_keywords), case=False)
]

# 정치 관련 키워드가 포함된 기사 제거
final_filtered = flood_related[
    ~flood_related['제목'].str.contains('|'.join(political_keywords), case=False) &
    ~flood_related['content'].str.contains('|'.join(political_keywords), case=False)
]

# 필터링된 결과 저장 (필요시)
final_filtered.to_csv('filtered_news_data.csv', index=False)

print("필터링 작업 완료! 결과는 'filtered_news_data.csv' 파일에 저장되었습니다.")
```

```python
import pandas as pd
import numpy as np

# CSV 파일 로드
file_path = 'NEW_뉴스_크롤링_data_clean.csv'
news_data = pd.read_csv(file_path)

# "침수" 관련 키워드 (이 키워드와 관계없이 마케팅 관련 키워드가 있으면 삭제)
flood_keywords = ['침수', '피해', '홍수', '물난리', '재해', '기상', '예보']

# 마케팅 관련 키워드 또는 침수와 무관한 키워드 (이 키워드가 들어가면 삭제)
irrelevant_keywords = ["세제", "나눔", "기부", "유통", "식품", "보험", "마케팅", "제품", "프로모션"]

# 특정 키워드가 등장하는 최소 횟수 설정 (예: 2회 이상)
min_keyword_count = 2

# 필터링 함수: 기사 내용에서 키워드 등장 횟수를 계산하고, 조건을 만족하는 기사만 남김
def keyword_count_filter(content, keywords, min_count):
    return sum(content.count(keyword) for keyword in keywords) >= min_count

# 1단계: 기사 내용에서 "침수" 관련 키워드가 최소 횟수 이상 등장하는 기사만 필터링
filtered_data = news_data[news_data['content'].apply(lambda x: keyword_count_filter(x, flood_keywords, min_keyword_count))]

# 2단계: 마케팅 관련 키워드 또는 불필요한 주제와 관련된 기사를 제거
# 불필요한 키워드가 포함된 모든 기사를 제거
filtered_data_strict = filtered_data[
    ~filtered_data['제목'].str.contains('|'.join(irrelevant_keywords), case=False, regex=True) &
    ~filtered_data['content'].str.contains('|'.join(irrelevant_keywords), case=False, regex=True)
]

# 3단계: 내용 길이 기반 필터링 (예: 100자 이하인 기사를 제거)
filtered_data_strict = filtered_data_strict[filtered_data_strict['content'].str.len() > 100]

# 필터링된 결과 저장
filtered_data_strict.to_csv('filtered_news_final.csv', index=False)

print("필터링 작업 완료! 결과는 'filtered_news_final.csv' 파일에 저장되었습니다.")
```

```python
import openai
import pandas as pd

# OpenAI API 키 설정
openai.api_key = ''

# CSV 파일 로드
file_path = 'NEW_뉴스_크롤링_data_clean.csv'
news_data = pd.read_csv(file_path)

# 정치 관련 키워드 설정
political_keywords = ["대통령", "의원", "시장", "정당", "청와대", "정부", "정치인", "총리"]

# 정치 관련 키워드가 포함된 기사를 제거하는 함수
def remove_political_articles(data):
    return data[
        ~data['제목'].str.contains('|'.join(political_keywords), case=False, regex=True) &
        ~data['content'].str.contains('|'.join(political_keywords), case=False, regex=True)
    ]

# 2. 프롬프트 생성 함수
def create_prompt(title, content):
    prompt = f"""
    다음은 뉴스 기사입니다:
    제목: {title}
    내용: {content[:200]}  # 기사 내용의 일부만 포함 (예: 첫 200자)

    이 뉴스가 "침수"와 관련된 뉴스인지 여부를 판단해 주세요. "예" 또는 "아니오"로만 답변해 주세요.
    """
    return prompt

# OpenAI API를 사용해 기사를 분류하는 함수
def classify_news(title, content):
    prompt = create_prompt(title, content)
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "당신은 뉴스 기사를 분류하는 전문가입니다."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=10,
        temperature=0
    )
    result = response['choices'][0]['message']['content'].strip()
    return result

# 샘플 데이터 추출
sample_data = news_data.sample(n=1000, random_state=42)

# 정치 관련 기사를 미리 제거
sample_data = remove_political_articles(sample_data)

# 샘플 데이터에 모델 적용
def filter_flood_related_news(data):
    flood_related = []

    for index, row in data.iterrows():
        title = row['제목']
        content = row['content']

        # 기사 분류
        classification = classify_news(title, content)

        # "예"라고 응답한 경우만 필터링
        if classification == "예":
            flood_related.append(row)

    return pd.DataFrame(flood_related)

# 필터링된 데이터 추출
filtered_sample_data = filter_flood_related_news(sample_data)

# 결과 확인
print(filtered_sample_data)
```
