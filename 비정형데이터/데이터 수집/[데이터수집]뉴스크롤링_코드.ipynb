```python
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wb
from selenium.webdriver.common.by import By

# 웹드라이버 초기화
driver = wb.Chrome()
driver.set_window_size(800, 600)

# URL 설정
Url = 'https://search.naver.com/search.naver?where=news&query=%EC%84%9C%EC%9A%B8%EC%8B%9C%20%EC%B9%A8%EC%88%98&sm=tab_opt&sort=2&photo=0&field=0&pd=3&ds=2018.01.01&de=2024.08.01&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom20180101to20240801&is_sug_officeid=0&office_category=0&service_area=0'
driver.get(Url)
time.sleep(6)

# 스크롤하여 전체 페이지 로드
last_height = driver.execute_script("return document.documentElement.scrollHeight")
all_urls = set()  # 중복된 뉴스 URL을 저장할 집합

while True:
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(1.5)
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    if new_height == last_height:
        break
    last_height = new_height
    time.sleep(1.5)
    
    try: 
        driver.find_element(By.CSS_SELECTOR, ".css-1topt76").click()
    except:
        pass
    
    soup = bs(driver.page_source, 'lxml') 
    
    title_list = soup.select(".news_tit")
    content_list = soup.select(".api_txt_lines.dsc_txt_wrap")
    comp_list = soup.select(".thumb_box")
    date_list = soup.select(".info")
    
    data = []

    for title, content, comp, date in zip(title_list, content_list, comp_list, date_list):
        title_text = title.text
        content_text = content.text
        comp_text = comp.text
        date_text = date.text
        
        
        
        if title_text not in all_urls:
            all_urls.add(title_text)
            data.append([title_text, content_text, comp_text, date_text])
            print(f'제목: {title_text}\n content: {content_text}\n언론사: {comp_text}\n날짜: {date_text}')

# 데이터프레임 생성 및 CSV 저장
df = pd.DataFrame(data, columns=['제목','content', '언론사', '날짜'])
df.to_csv('flooding_naver_crawling.csv', index=False, encoding='utf-8-sig')

# 웹드라이버 종료
driver.quit()
```

```python
## main
## main
## main

from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup as bs
import tqdm
import pandas as pd

# WebDriver 설정
driver = webdriver.Chrome()
driver.get("https://search.naver.com/search.naver?where=news&query=%EC%84%9C%EC%9A%B8%EC%8B%9C%20%EC%B9%A8%EC%88%98&sm=tab_opt&sort=2&photo=0&field=0&pd=3&ds=2024.07.09&de=2024.08.14&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom20240709to20240814&is_sug_officeid=0&office_category=0&service_area=0")

# 스크롤 높이의 3분의 2만큼 스크롤
last_height = driver.execute_script("return document.documentElement.scrollHeight")

data = []
collected_comments = set()  # 수집된 댓글을 저장하기 위한 집합

# tqdm progress bar 설정
max_scrolls = 500  # 임시로 최대 스크롤 횟수 설정
with tqdm.tqdm(total=max_scrolls) as pbar:
    scroll_count = 0
    while True:
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(1.5)
        new_height = driver.execute_script("return document.documentElement.scrollHeight")
        
        # 현재 페이지의 HTML 소스를 다시 가져옴
        soup = bs(driver.page_source, 'lxml')

        # 뉴스 기사 선택
        articles = soup.select('.news_area')

        new_comments = 0  
        
        for article in articles:
            title_element = article.select_one('a.news_tit')
            content_element = article.select_one('.api_txt_lines.dsc_txt_wrap')
            comp_element = article.select_one('div.news_info > div.info_group > a')
            date_element = article.select_one('.info_group > span')

            if title_element and content_element and comp_element and date_element:
                title_text = title_element.text.strip()
                content_text = content_element.text.strip()
                comp_text = comp_element.text.strip()
                date_text = date_element.text.strip()
                link = title_element.get('href')

                # 중복 확인
                if content_text not in collected_comments:
                    collected_comments.add(content_text)
                    data.append([title_text, content_text, comp_text, date_text, link])
                    new_comments += 1
                    # print(f'제목: {title_text}\ncontent: {content_text}\n언론사: {comp_text}\n날짜: {date_text}\n')
                    time.sleep(1)

        # 새로운 댓글이 없으면 스크래핑 종료
        if new_comments == 0:
            print("더 이상 새로운 댓글이 없습니다. 스크래핑을 종료합니다.")
            break
        
        scroll_count += 1
        pbar.update(1)
        
        # 페이지 로딩 횟수 제한 (무한 스크롤 방지)
        if scroll_count >= max_scrolls:
            print("스크롤 횟수 제한에 도달했습니다. 스크래핑을 종료합니다.")
            break

# DataFrame으로 변환하여 CSV로 저장
df = pd.DataFrame(data, columns=['제목', 'content', '언론사', '날짜', '링크'])
df.to_csv('flooding_naver_crawling_22.csv', index=False, encoding='utf-8-sig')
```
